```python
import pandas as pd
project_df_config = pd.read_csv('/content/ProjectConfiguration.csv', sep=',', on_bad_lines='skip')
display(project_df_config.head())
```

```python
project_df_config_var = pd.read_csv('/content/ProjectConfigurationVariant.csv', sep=',', on_bad_lines='skip')
display(project_df_config_var.head())
```

```python
# Clean up column names by removing quotes and spaces
project_df_config_var.columns = project_df_config_var.columns.str.replace('"', '').str.strip()
project_df_config.columns = project_df_config.columns.str.strip()

# Perform the left merge
merged_config_df = pd.merge(project_df_config_var, project_df_config, left_on='configurationId', right_on='id', how='left')

# Display the first few rows of the merged dataframe
display(merged_config_df.head())
```

```python
print(f"The merged_config_df has {merged_config_df.shape[0]} rows and {merged_config_df.shape[1]} columns.")
```

```python
merged_config_df.to_csv('final_merged_data100.csv', index=False)
```

```python
project_adr_df = pd.read_csv('/content/ProjectAddress.csv', sep=',', on_bad_lines='skip')
display(project_adr_df.head())
```

```python
print("Columns in merged_config_df:", merged_config_df.columns)
print("Columns in project_adr_df:", project_adr_df.columns)
```

```python
# Clean up column names by removing quotes and spaces
merged_config_df.columns = merged_config_df.columns.str.strip()
project_adr_df.columns = project_adr_df.columns.str.strip()

# Perform the left merge
final_merged_df = pd.merge(merged_config_df, project_adr_df, on='projectId', how='left')

# Display the first few rows of the merged dataframe
display(final_merged_df.head())
```

```python
print(f"The final_merged_df has {final_merged_df.shape[0]} rows and {final_merged_df.shape[1]} columns.")
```

```python
final_merged_df.to_csv('final_merged_data200.csv', index=False)
```
